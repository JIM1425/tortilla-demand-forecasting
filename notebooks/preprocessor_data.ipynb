## Preprocesamiento de datos

```python
import pandas as pd
import numpy as np
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("TC_Data_synthetic.csv")
```

```python
df.head()
```

```python
df.dropna(subset=["fecha"], inplace=True)
```

```python
df["repartidor"] = df["repartidor"].fillna("NA")
```

```python
df["entrega"] = df["entrega"].fillna(0)
```

```python
df.isna().sum()
```

```python
df['cliente_id'], unique_clientes = pd.factorize(df['cliente'])
```

```python
df.drop(columns=["cliente"], inplace=True)
```

```python
df["fecha"] = df["fecha"].astype("datetime64[ns]")
```

```python
df["dia"] = df["fecha"].dt.isocalendar().day
```

```python
df["mes"] = df["fecha"].dt.month
```

```python
def estaciones(mes):
    if mes in [12, 1, 2]:
        return 1  # Invierno
    elif mes in [3, 4, 5]:
        return 2  # Primavera
    elif mes in [6, 7, 8]:
        return 3  # Verano
    else:
        return 4  # Otoño

df["estacion"] = df["mes"].apply(estaciones)
```

```python
columna = df.pop('cliente_id')
df.insert(0, 'cliente_id', columna)
```

```python
df['repartidor_id'], unique_repartidores = pd.factorize(df['repartidor'])
columna = df.pop('repartidor_id')
df.insert(5, 'repartidor_id', columna)
```

```python
df.drop(["repartidor"], axis=1, inplace=True)
```

```python
df["año"] = df["fecha"].dt.year
df["dia_mes"] = df["fecha"].dt.day
```

```python
df.drop(["fecha"], axis=1, inplace=True)
```

```python
df["dia_sem"] = df["dia"]
df.drop(["dia"], axis=1, inplace=True)
```

```python
df["demanda_neta"] = df["entrega"] - df["devolución"]  # Variable determinante

df.drop(["masa_tortilla", "masa_tamal" ], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"devolución" : "devolucion"}, inplace=True)
```

```python
df.head()
```

```python
df.to_csv("data_synthetic.csv")
```
