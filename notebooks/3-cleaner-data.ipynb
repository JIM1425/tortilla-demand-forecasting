```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("Enero_2021/5-1-_21.xlsx",skiprows=7)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.shape
```

```python
df.drop(columns={"Total","No pagó","No percibido","Unnamed: 10", "Unnamed: 11","Unnamed: 12","Unnamed: 13","Unnamed: 14"},  inplace=True)
```

```python
df.head(100)
```

```python
df.rename(columns={"M/Tortilla":"Masa_tortilla","M/Tamales":"Masa_tamal"},inplace=True)
```

```python
df["Repartidor"]=""
```

```python
df["REP"].value_counts()
```

```python
df["REP"]=df["REP"].astype(str)
```

```python
for i in df.index:
    if df.loc[i,"REP"]=="D":
        df.loc[i,"Repartidor"]="Diego"
    elif df.loc[i,"REP"]=="L":
        df.loc[i,"Repartidor"]="Laura"
    elif df.loc[i,"REP"]=="A":
        df.loc[i,"Repartidor"]="Alex"
    else:
        df.loc[i,"Repartidor"]="NA"
```

```python
df["Repartidor"].value_counts()
```

```python
df.drop(columns={"REP"},inplace=True)
```

```python
df["Fecha"]="05-01-2021"
```

```python
df["Fecha"]=pd.to_datetime(df["Fecha"], format='%d-%m-%Y')
```

```python
df["Entrega"]=df["Entrega"].fillna(0)
df["Devolución"]=df["Devolución"].fillna(0)
df["Masa_tortilla"]=df["Masa_tortilla"].fillna(0)
df["Masa_tamal"]=df["Masa_tamal"].fillna(0)
df["Sin epecificar"]=df["Sin epecificar"].fillna(0)
```

```python
df.dropna(axis=0, how="any",subset="Cliente",inplace=True)
```

```python
df.drop({62,69},  inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df_1 = df.iloc[:35].copy()
df_2 = df.iloc[35:].copy()
```

```python
df_1.drop(columns={"Sin epecificar"},inplace=True)
```

```python
df_2.drop(columns={"Devolución"},inplace=True)
```

```python
df_2.rename(columns={"Masa_tortilla":"Devolución","Masa_tamal":"Masa_tortilla","Sin especificar":"Masa_tamal"})
```

```python
df_f = pd.DataFrame(np.vstack([df_1.to_numpy(), df_2.to_numpy()]), columns=df_1.columns)
```

```python
df_f.head(100)
```

```python
df_f.drop({35,36}, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df_f["Cliente"]=df_f["Cliente"].astype(str)
df_f["Entrega"]=df_f["Entrega"].astype(float)
df_f["Devolución"]=df_f["Devolución"].astype(float)
df_f["Masa_tortilla"]=df_f["Masa_tortilla"].astype(float)
df_f["Masa_tamal"]=df_f["Masa_tamal"].astype(float)
df_f["Repartidor"]=df_f["Repartidor"].astype(str)
```

```python
df_f.head(100)
```

```python
df_f.to_csv("cleaned_enero_2021/05-01-2021.xlsx",index=False)
```
